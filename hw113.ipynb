```python
# Hunter Huynh
# Helps with ECE 470 HW 1.1.3

###############################################################################

# The target rotation matrix
# Input the rotation matrix here:
target_row_1 = [0.5251, 0.1404, 0.8393]
target_row_2 = [0.5853, 0.6563, -0.4760]
target_row_3 = [-0.6177, 0.7413, 0.2625]
target_matrix = [target_row_1, target_row_2, target_row_3]

###############################################################################

import math
import numpy as np
def c(x): return math.cos(x)
def s(x): return math.sin(x)

# Input: Euler angles (phi, theta, psi) in degrees
# Output: Rotation matrix as a result of:
#   - rotating about the z-axis by the angle phi,
#   - then, rotating about the current y-axis by the angle theta,
#   - then, rotating about the current z-axis by the angle psi.
def get_rotation_matrix(phi, theta, psi):
    p = math.radians(phi)
    t = math.radians(theta)
    v = math.radians(psi)
    row_1 = [c(p)*c(t)*c(v) - s(p)*s(v), -c(p)*c(t)*s(v) - s(p)*c(v), c(p)*s(t)]
    row_2 = [s(p)*c(t)*c(v) + c(p)*s(v), -s(p)*c(t)*s(v) + c(p)*c(v), s(p)*s(t)]
    row_3 = [-s(t)*c(v), s(t)*s(v), c(t)]
    rot_matrix = [row_1, row_2, row_3]
    return rot_matrix

# Input: target rotation matrix
# Outputs: none
# Side Effects: prints Euler angles in degrees
def find_euler_angles(target_matrix):
    print("Target matrix:", target_matrix)
    max_error = 0.2
    while True:
        print("Searching...")
        for i in range(0, 360, 5):
            for j in range(0, 360, 5):
                for k in range(0, 360, 5):
                    result_rotation_matrix = get_rotation_matrix(i, j, k)
                    error_matrix = np.fabs(np.array(result_rotation_matrix) -
                        np.array(target_matrix))
                    error = np.sum(error_matrix)
                    if error < max_error:
                        print("\nSuccess!")
                        print("Euler angles: phi =", i, "theta =", j, "psi =", k)
                        print("Instructions:")
                        if i > 180:
                            i = -(360 - i)
                            print("1. Rotate about z by", i, "degrees (press < z button", int(-i / 5), "times)")
                        else:
                            print("1. Rotate about z by", i, "degrees (press z > button", int(i / 5), "times)")
                        if j > 180:
                            j = -(360 - j)
                            print("2. Rotate about y by", j, "degrees (press < y button", int(-j / 5), "times)")
                        else:
                            print("2. Rotate about y by", j, "degrees (press y > button", int(j / 5), "times)")
                        if k > 180:
                            k = -(360 - k)
                            print("3. Rotate about z by", k, "degrees (press < z button", int(-k / 5), "times)")
                        else:
                            print("3. Rotate about z by", k, "degrees (press z > button", int(k / 5), "times)")
                        return
        max_error += 0.1
        if max_error > 2:
            break
    print("\nFailure. Could not find a solution. :(")

find_euler_angles(target_matrix)
```
